```python
from IPython.display import HTML
import random

def hide_toggle(for_next=False):
    this_cell = """$('div.cell.code_cell.rendered.selected')"""
    next_cell = this_cell + '.next()'

    toggle_text = 'Toggle show/hide'  # text shown on toggle link
    target_cell = this_cell  # target cell to control with toggle
    js_hide_current = ''  # bit of JS to permanently hide code in current cell (only when toggling next cell)

    if for_next:
        target_cell = next_cell
        toggle_text += ' next cell'
        js_hide_current = this_cell + '.find("div.input").hide();'

    js_f_name = 'code_toggle_{}'.format(str(random.randint(1,2**64)))

    html = """
        <script>
            function {f_name}() {{
                {cell_selector}.find('div.input').toggle();
            }}

            {js_hide_current}
        </script>

        <a href="javascript:{f_name}()">{toggle_text}</a>
    """.format(
        f_name=js_f_name,
        cell_selector=target_cell,
        js_hide_current=js_hide_current, 
        toggle_text=toggle_text
    )

    return HTML(html)

hide_toggle()
```

```python
# -*- coding: utf-8 -*-
"""
Created on Fri Oct 29 13:13:17 2021

@author: Sami
"""

import numpy as np
import torch
import random


class Task():

    def __init__(self, n_hidden_features,curve1=[],curve2=[]):

        self.grid_size = 18
        
        self.bias = False

        self.state = 0
        self.trial_ended = False

        self.only_trace_curve = True # For the first part of the training of the trace then search task, we only do the curve tracing task

        self.state = 'intertrial'

        self.no_curves = True # First part of the curicculum is no curve, only the fixation point and a blue pixel

        self.current_reward = 0
        self.final_reward = 1.5
        
        self.curve1 = curve1
        self.curve2 = curve2
        
        self.n_hidden_features = n_hidden_features

        self.flowcontrol = {'intertrial': self.do_intertrial,
                            'go': self.do_go}

        if self.task_type != 'searchtrace' and self.task_type != 'trace' and self.task_type != 'tracesearch':
            raise Exception('Task type must be either searchtrace or trace or tracesearch')

    def stateReset(self):
        self.trial_ended = True
        self.state = 'intertrial'
        self.input = [self.display, self.display_disk]

    def doStep(self, action):
        self.trial_ended = False
        self.flowcontrol[self.state](action)
        reward = self.current_reward
        input = self.input
        trial_ended = self.trial_ended
        self.current_reward = 0
        return(input, reward, trial_ended)

    def do_intertrial(self, action):
            self.pickTrialType()
            self.input = [self.display, self.display_disk]
            self.state = 'go'


    def do_go(self, action):
            if (self.task_type == 'tracesearch' and self.only_trace_curve is False):
                pixel_chosen = torch.where(action == 1)[-1]
            else:
                pixel_chosen = torch.where(action == 1)[-1]
            if pixel_chosen == self.target_curve[-1]:
                self.current_reward = self.current_reward + self.final_reward * 0.8
                self.stateReset()
            else:
                self.stateReset()

    def pickTrialType(self):
        position_red_marker = np.random.randint(2)
        position_yellow_marker = 1 if position_red_marker == 0 else 0
        self.position_markers = [position_red_marker, position_yellow_marker]
        self.feature_target = np.random.randint(2) 
        curve1,curve2 = self.PickCurve()
        self.DrawStimulus(curve1,curve2)
        
    def PickCurve(self):
        if self.no_curves:
            red_yellow_position = np.random.randint(self.grid_size**2)
            blue_position = np.random.randint(self.grid_size**2)
            while blue_position == red_yellow_position:  ########CHANGE THAT
                blue_position = np.random.randint(self.grid_size**2)
            curve1 = [red_yellow_position, blue_position]
            curve2 = []
        else:
          if len(self.curve1) == 0:
            while True:
                try:
                    mask = np.ones((self.grid_size, self.grid_size))
                    mask[:,:8] = 0
                    flag = True
                    while flag:
                      curve1, mask1 = self.make_curves([], mask)
                      curve1Big = [((curve1[i] // self.grid_size) // 3)*6 + (curve1[i] % self.grid_size) // 3 for i in range(len(curve1))]
                      curve1BigUnique = np.unique(curve1Big)
                      if curve1BigUnique == curve1Big:
                        flag = False
                      else:
                        position = [np.where(curve1Big == curve1BigUnique[i]) for i in range(len(curve1BigUnique))]
                        consecutive_boolean = [sorted(position[i][0]) == list(range(min(position[i][0]), max(position[i][0])+1)) for i in range(len(position))]
                        flag = False in consecutive_boolean
                    #curve2, mask2 = self.make_curves([], mask1)
                    curve2 = []
                    for i in range(len(curve1)): #trnaspose
                        x = curve1[i] % self.grid_size
                        y = 9 + curve1[i]//self.grid_size
                        curve2.append(x + y*self.grid_size)    
                    break
                except IndexError:
                    pass
          else:
                curve1 = self.curve1
                curve2 = self.curve2
        return (curve1, curve2)
                
    def make_curves(self,curve, mask_original):
        mask = mask_original.copy()
        if len(curve) == 0:
            # Getting the pixels available
            x, y = np.where(mask == 0)
            
            # Chossing a random pixel among the one availables
            ind = np.random.randint(len(x))
            first_elem = x[ind] + y[ind] * self.grid_size
            
            #We mask this pixel, and call the function again
            mask[first_elem % self.grid_size, first_elem // self.grid_size] = 1
            curve.append(first_elem)
            return self.make_curves(curve, mask)
        else:
            # We look at the last pixel of the curve and get the coordinate of 
            # its neihbours
            x_end = curve[-1] % self.grid_size
            y_end = curve[-1] // self.grid_size
            consecutives_x = [x_end - 1, x_end + 1]
            consecutives_x = [i for i in consecutives_x if i >= 0 and i < self.grid_size]
            consecutives_y = [y_end - 1, y_end + 1]
            consecutives_y = [i for i in consecutives_y if i >= 0 and i < self.grid_size]
            
            while len(curve) < self.curve_length:  
                possible_next_value = []
                
                # If a neighbouring pixel is available, we record it as a
                # a possible value
                for i in range(len(consecutives_x)):
                    if mask[consecutives_x[i], y_end] == 0:
                        possible_next_value.append((consecutives_x[i], y_end))
                for i in range(len(consecutives_y)):
                    if mask[x_end, consecutives_y[i]] == 0:
                        possible_next_value.append((x_end, consecutives_y[i]))
                
                # We randomly choose the next pixel among the available ones
                # and mask the nehbours 
                weights = [1/len(possible_next_value) for i in range(len(possible_next_value))]
                if self.bias and len(curve) > 1:
                    last_2_x = [curve[-i] % self.grid_size for i in range(1,3)]
                    last_2_y = [curve[-i] // self.grid_size for i in range(1,3)]
                    if last_2_x.count(last_2_x[0]) == len(last_2_x): #alignement of x
                        for p in range(len(possible_next_value)):
                            if possible_next_value[p][0] == last_2_x[0]:
                                weights[p] = 100

                    if last_2_y.count(last_2_y[0]) == len(last_2_y): #alignement of y
                        for p in range(len(possible_next_value)):
                            if possible_next_value[p][1] == last_2_y[1]:
                                weights[p] = 100
                if weights.count(weights[0]) != len(weights):
                    s = sum(weights)
                    for p in range(len(weights)):
                        weights[p] = weights[p]/s
                next_value = random.choices(possible_next_value,weights=weights)[0]
                curve.append(next_value[0] + next_value[1] * self.grid_size)
                for i in range(len(possible_next_value)):
                    mask[possible_next_value[i][0], possible_next_value[i][1]] = 1
                return self.make_curves(curve, mask)
            else:
                
                # If we reached the length of the curve, we only mask the nehbours
                for i in range(len(consecutives_x)):
                    mask[consecutives_x[i], y_end] = 1
                for i in range(len(consecutives_y)):
                    mask[x_end, consecutives_y[i]] = 1
                return curve, mask

        
class Trace(Task):
    
    def __init__(self,n_hidden_features,curve1=[],curve2=[]):
        self.task_type = 'trace'
        super().__init__(n_hidden_features,curve1,curve2)
        
    def DrawStimulus(self,curve1,curve2):
        display = torch.zeros((1, self.n_hidden_features, self.grid_size, self.grid_size))
        display_disk = torch.zeros((1, self.n_hidden_features, 2))
        if self.no_curves:
                display[:, 0, curve1[0] % self.grid_size, curve1[0]//self.grid_size] = 0
                display[:, 3, curve1[1] % self.grid_size, curve1[1]//self.grid_size] = 1
        else:
                for i in range(len(curve1)):
                    if i == 0:  # red
                        display[:, 0, curve1[0] % self.grid_size, curve1[0]//self.grid_size] = 1
                    elif i == len(curve1)-1:  # blue
                        display[:, 3, curve1[i] % self.grid_size, curve1[i]//self.grid_size] = 1
                    else:  # green
                        display[:, 2, curve1[i] % self.grid_size, curve1[i]//self.grid_size] = 1
                for i in range(len(curve2)):
                    if i == 0:
                        display[:, 2, curve2[0] % self.grid_size, curve2[0]//self.grid_size] = 1
                    elif i == len(curve2)-1:
                        display[:, 3, curve2[i] % self.grid_size, curve2[i]//self.grid_size] = 1
                    else:
                        display[:, 2, curve2[i] % self.grid_size, curve2[i]//self.grid_size] = 1

        self.target_curve = curve1
        self.distractor_curve = curve2
        self.display = display
        self.display_disk = display_disk     
        
class SearchTrace(Task):
    
    def __init__(self,n_hidden_features):
        self.task_type = 'searchtrace'
        super().__init__(n_hidden_features)
        
    def DrawStimulus(self,curve1,curve2):
        display = torch.zeros((1, self.n_hidden_features, self.grid_size, self.grid_size))
        display_disk = torch.zeros((1, self.n_hidden_features, 2))
        if self.no_curves:
          display[:, self.feature_target, curve1[0] % self.grid_size, curve1[0]//self.grid_size] = 1
          display[:, 3, curve1[1] % self.grid_size, curve1[1]//self.grid_size] = 1
          display_disk[:, self.feature_target, self.position_markers[self.feature_target]] = 1 
        else:
          for i in range(len(curve1)):
              if i == 0:  # red
                  display[:, 0, curve1[0] % self.grid_size, curve1[0]//self.grid_size] = 1
              elif i == len(curve1)-1:  # blue
                  display[:, 3, curve1[i] % self.grid_size, curve1[i]//self.grid_size] = 1
              else:  # green
                  display[:, 2, curve1[i] % self.grid_size, curve1[i]//self.grid_size] = 1
          for i in range(len(curve2)):
              if i == 0:
                  display[:, 1, curve2[0] % self.grid_size, curve2[0]//self.grid_size] = 1
              elif i == len(curve2)-1:
                  display[:, 3, curve2[i] % self.grid_size, curve2[i]//self.grid_size] = 1
              else:
                  display[:, 2, curve2[i] % self.grid_size, curve2[i]//self.grid_size] = 1
          display_disk[:, self.feature_target, self.position_markers[self.feature_target]] = 1
        if self.no_curves:
            self.target_curve = curve1
            self.distractor_curve = curve2
        else:
            if self.feature_target == 0:
                self.target_curve = curve1
                self.distractor_curve = curve2
            else:
                self.target_curve = curve2
                self.distractor_curve = curve1
        self.display = display
        self.display_disk = display_disk
        
class TraceSearch(Task):      
    
    def __init__(self,n_hidden_features):
        self.task_type = 'tracesearch'
        super().__init__(n_hidden_features)
        
    def DrawStimulus(self,curve1,curve2):
        
        # The eye movement target is the red or yellow pixel so we reverse the
        # curves
        curve1.reverse()
        curve2.reverse()
        if not self.only_trace_curve:
            if self.no_curves:
                curve1.append(self.position_markers[self.feature_target])
            else:
                curve1.append(self.position_markers[0])
                curve2.append(self.position_markers[1])       
        display = torch.zeros((1, self.n_hidden_features, self.grid_size, self.grid_size))
        display_disk = torch.zeros((1, self.n_hidden_features, 2))
        if self.no_curves:
            display[:, 3, curve1[0] % self.grid_size, curve1[0]//self.grid_size] = 1
            display[:, self.feature_target, curve1[1] % self.grid_size, curve1[1]//self.grid_size] = 1
            if not self.only_trace_curve:
                display_disk[:, 0, self.position_markers[0]] = 1
                display_disk[:, 1, self.position_markers[1]] = 1
        else:
            if self.only_trace_curve:
                display[:, 0, curve1[len(curve1)-1] % self.grid_size, curve1[len(curve1)-1]//self.grid_size] = 1
                display[:, 1, curve2[len(curve2)-1] % self.grid_size, curve2[len(curve2)-1]//self.grid_size] = 1
                range1 = len(curve1) - 1
                range2 = len(curve2) - 1
            else:
                display[:, 0, curve1[len(curve1)-2] % self.grid_size, curve1[len(curve1)-2]//self.grid_size] = 1
                display[:, 1, curve2[len(curve2)-2] % self.grid_size, curve2[len(curve2)-2]//self.grid_size] = 1
                range1 = len(curve1) - 2
                range2 = len(curve2) - 2
            for i in range(range1):
                display[:, 2, curve1[i] % self.grid_size, curve1[i]//self.grid_size] = 1
            for i in range(range2):
                display[:, 2, curve2[i] % self.grid_size, curve2[i]//self.grid_size] = 1
            if self.feature_target == 0:
                display[:, 3, curve1[0] % self.grid_size, curve1[0]//self.grid_size] = 1
                display[:, 2, curve1[0] % self.grid_size, curve1[0]//self.grid_size] = 0
            else:
                display[:, 3, curve2[0] % self.grid_size, curve2[0]//self.grid_size] = 1
                display[:, 2, curve2[0] % self.grid_size, curve2[0]//self.grid_size] = 0
            if not self.only_trace_curve:
                display_disk[:, 0, self.position_markers[0]] = 1
                display_disk[:, 1, self.position_markers[1]] = 1
        if self.no_curves:
            self.target_curve = curve1
            self.distractor_curve = curve2
        else:
            if self.feature_target == 0:
                self.target_curve = curve1
                self.distractor_curve = curve2
            else:
                self.target_curve = curve2
                self.distractor_curve = curve1
        self.display = display
        self.display_disk = display_disk
    
        
    
```

```python
# -*- coding: utf-8 -*-
"""
Created on Fri Oct 29 12:29:28 2021

@author: Sami
"""
import numpy as np
import torch
import torch.nn as nn


class CustomLayer(nn.Module):

    def __init__(self):
        self.initialisation_range = 0.1
        super().__init__()

    def ActivationFunction(self, x):
        thresh = 6
        x[x == 0] = 0
        x = torch.clamp(x, 0)
        x[thresh <= x] = torch.log(1.5 * x[thresh <= x] + 1) + thresh - torch.log(torch.tensor(1.5 * thresh + 1))
        return x

    def MultiFunction(self,x):
      param = 100
      return param*x/(torch.sqrt(1+(param**2)*(x**2)))
      #return x

    def AverageTraces(self, traces, mask,skip=False,change_scale=False):
        traces *= mask
        if self.LayerType != "output":
            if change_scale:
                m = torch.mean(traces[:,:,:,:], axis=(2,3))
                m = m[:, :, None,None]
                traces[:,:] = m  
            else:
                m = torch.mean(traces[:, :, [0, 1, 1, 2], [1, 0, 2, 1]], axis=2)
                m = m[:, :, None]
                traces[:, :, [0, 1, 1, 2], [1, 0, 2, 1]] = m
        elif self.LayerType == "output" and not skip:
            traces *= mask
            intermmask = torch.ones((2 * self.grid_size - 1, 2 * self.grid_size - 1))
            intermmask[16:19,16:19] = 0
            m = torch.mean(traces[:, :, intermmask == 1], axis=2)
            traces[:, :, intermmask == 1] =  m[:,None]

            intermmask = torch.zeros((2 * self.grid_size - 1, 2 * self.grid_size - 1))
            intermmask[16:19,16:19] = 1
            m = torch.mean(traces[:, :, intermmask == 1], axis=2)
            traces[:, :, intermmask == 1] =  m[:,None]            
        elif self.LayerType == 'output' and skip:
            traces *= mask
            intermmask = torch.ones((2 * self.grid_size - 1, 2 * self.grid_size - 1))
            intermmask[self.grid_size - 1, self.grid_size - 1] = 0
            m = torch.mean(traces[:, :, intermmask == 1], axis=2)
            traces[:, :, intermmask == 1] = m.T          
        return(traces)

    def init_weights_FF(self, layer, disk=False,diag=False,change_scale=False,higher_scale = False,skip=False,rangee=1):
        K = torch.zeros(layer.weight.shape)
        if layer.bias is not None:
            bias = torch.zeros(layer.bias.shape)
        if change_scale:
          for f in range(K.shape[1]): 
            K[0,f,:,:] = torch.tensor([[[0.1* np.random.rand()+0.2,-1,-1],[0.1* np.random.rand()+0.2,-1,-1],[0.1* np.random.rand()+0.2,-1,-1]]])
            K[1,f,:,:] = torch.tensor([[[-1,0.1* np.random.rand()+0.2,-1],[-1,0.1* np.random.rand()+0.2,-1],[-1,0.1* np.random.rand()+0.2,-1]]])
            K[2,f,:,:] = torch.tensor([[[-1,-1,0.1* np.random.rand()+0.2],[-1,-1,0.1* np.random.rand()+0.2],[-1,-1,0.1* np.random.rand()+0.2]]])
            K[3,f,:,:] = torch.tensor([[[0.1* np.random.rand()+0.2,0.1* np.random.rand()+0.2,0.1* np.random.rand()+0.2],[-1,-1,-1],[-1,-1,-1]]])
            K[4,f,:,:] = torch.tensor([[[-1,-1,-1],[0.1* np.random.rand()+0.2,0.1* np.random.rand()+0.2,0.1* np.random.rand()+0.2],[-1,-1,-1]]])
            K[5,f,:,:] = torch.tensor([[[-1,-1,-1],[-1,-1,-1],[0.1* np.random.rand()+0.2,0.1* np.random.rand()+0.2,0.1* np.random.rand()+0.2]]])
          if layer.bias is not None:
              bias[f] = 0.1 * np.random.rand()
        if higher_scale and not change_scale:
          for f in range(K.shape[0]):
              for f2 in range(K.shape[1]):
                if f == f2 :
                      K[f, f2, 1, 1] = 0.1 * np.random.rand() + 0.2                                
        if not higher_scale:
          for f in range(K.shape[0]):
              for f2 in range(K.shape[1]):
                  if not disk:
                      if self.LayerType != "output":
                          K[f, f2, 1, 1] = self.initialisation_range * np.random.rand() * rangee
                      else:
                          if skip == False:
                              K[f, f2, :, :] = - 0.1 * np.random.rand() * torch.ones((K.shape[2], K.shape[3])) / 100
                              K[f, f2, 16:19,16:19] = 0.1 * np.random.rand()
                              #K[f, f2, :, :] = 0.1 * np.random.rand()
                          else:
                              K[f, f2, :, :] =  -0.1 * np.random.rand() * torch.ones((K.shape[2], K.shape[3])) / 100
                              K[f, f2, self.grid_size - 1, self.grid_size - 1] = 0.1 * np.random.rand()       
                  else:
                      K[f, f2, 0] = self.initialisation_range * np.random.rand()
              if layer.bias is not None:
                for f in range(bias.shape[0]):   
                  bias[f] = 0.1 * np.random.rand()
        layer.weight = torch.nn.Parameter(K)
        if layer.bias is not None:
            layer.bias = torch.nn.Parameter(bias)

    def init_weights_FB(self, layer, disk=False,diag=False,change_scale=False,higher_scale = False):
        K = torch.zeros(layer.weight.shape)
        if layer.bias is not None:
            bias = torch.zeros(layer.bias.shape)
        if higher_scale:
          for f in range(K.shape[0]):
              for f2 in range(K.shape[1]):
                if f == f2 :
                    if f2 in [0,1,2]:
                          K[f, f2, 1, 1] = 0.1 * np.random.rand()
                          #K[f, f2, [0, 1, 1, 2], [1, 0, 2, 1]] = 0.1 * np.random.rand()
                          K[f, f2, 0,1] = 0.1 * np.random.rand()+ 0.2
                          K[f, f2, 2,1] = 0.1 * np.random.rand()+ 0.2
                    elif f2 in [3,4,5]:
                          K[f, f2, 1, 1] = 0.1 * np.random.rand()
                          #K[f, f2, [0, 1, 1, 2], [1, 0, 2, 1]] = 0.1 * np.random.rand()
                          K[f, f2, 1,0] = 0.1 * np.random.rand()+ 0.2
                          K[f, f2, 1,2] = 0.1 * np.random.rand()+ 0.2                                                 
        else:                                            
          for f in range(K.shape[0]):
              for f2 in range(K.shape[1]):
                  if not disk:
                      if self.LayerType != "output":
                        if change_scale:
                          K[f, f2, :, :] = 0.1 * np.random.rand() + 0.2
                        else:
                            K[f, f2, 1, 1] = 0.1 * np.random.rand() 
                            #K[f, f2, [0, 1, 1, 2], [1, 0, 2, 1]] = 0.1 * np.random.rand()
                            K[f, f2, 0,1] = 0.1 * np.random.rand()+ 0.2
                            K[f, f2, 1,0] = 0.1 * np.random.rand()+ 0.2
                            K[f, f2, 1,2] = 0.1 * np.random.rand()+ 0.2
                            K[f, f2, 2,1] = 0.1 * np.random.rand()+ 0.2
                  else:
                      K[f, f2, 0] = self.initialisation_range * np.random.rand()
        if layer.bias is not None:
          for f in range(bias.shape[0]):   
            bias[f] = 0.1 * np.random.rand()
        layer.weight = torch.nn.Parameter(K)
        if layer.bias is not None:
            layer.bias = torch.nn.Parameter(bias)


    def UpdateWeight(self, layer, upper, beta, delta, mask=None, z=None, average=True,change_scale=False,skip=False):
        with torch.no_grad():
            
            # Getting the derivative of the output unit with respect to the 
            # weight and bias
            if layer.bias is not None:
                delta_weight = torch.autograd.grad(upper, [layer.weight, layer.bias], grad_outputs=z, retain_graph=True, allow_unused=True)
                delta_bias = delta_weight[1]
                delta_bias = torch.clamp(delta_bias, None, 1)
                #delta_bias[:] = torch.mean(delta_bias)
            else:
                delta_weight = torch.autograd.grad(upper, layer.weight, grad_outputs=z, retain_graph=True, allow_unused=True)
            delta_weight = delta_weight[0]
            delta_weight = torch.clamp(delta_weight, None, 1)
            
            # Averaging the weight and bias and updating the weights with RPE
            if average:
              if self.LayerType == 'output' or self.change_scale_u:
                delta_weight = self.AverageTraces(delta_weight, mask,change_scale=change_scale,skip=skip)
              else:
                delta_weight = delta_weight*mask
            weight_update = layer.weight + beta * delta * delta_weight
            layer.weight.copy_(weight_update)
            if layer.bias is not None:
                bias_update = layer.bias + beta * delta * delta_bias
                layer.bias.copy_(bias_update)
        return(layer)

    def to(self, device):
        for i in range(2):
            if self.LayerType == 'input':
                self.u[i].to(device)
                self.umod[i].to(device)
                self.umask = self.umask.to(device)
            elif self.LayerType == 'hidden':
                self.v[i].to(device)
                self.t[i].to(device)
                self.vtmask = self.vtmask.to(device)
                if self.has_Ymod:
                    self.u[i].to(device)
                    self.umask = self.umask.to(device)
                    if self.upper_ymod:
                        self.umod[i].to(device)
            elif self.LayerType == 'output':
                self.w[i].to(device)
                self.skip[i].to(device)
                self.wmask = self.wmask.to(device)
                self.skipmask = self.skipmask.to(device)
                
                
class InputLayer(CustomLayer):

    def __init__(self, feature_in, feature_out):
        super().__init__()
        self.LayerType = 'input'
        K_size = 3
        self.change_scale_u = False
        # First element is grid to grid, second element is disk to disk
        self.u = nn.Conv2d(feature_out, feature_in, K_size, stride=1, padding='same')
        self.umod = nn.Conv2d(feature_out, feature_in, K_size, stride=1, padding='same',bias=False)

        # Initializing weights
        self.init_weights_FB(self.u)
        self.init_weights_FB(self.umod)

        # Setting the mask to have connections only between neighboours
        self.umask = torch.tensor([[0, 1, 0], [1, 1, 1], [0, 1, 0]], requires_grad=False)
        self.umask = torch.tile(self.umask, (feature_in, feature_out, 1, 1))

    def forward(self, upper_y, upper_ymod, input):
        Y = input[0]
        return(Y)

    def backward(self,same_y, upper_y, upper_ymod, input):        
        Ymod = self.ActivationFunction(self.umod(upper_ymod) * self.MultiFunction(same_y))
        return(Ymod)

    def UpdateLayer(self, upper, z, beta, delta):
        #self.u= self.UpdateWeight(self.u, upper, beta, delta, self.umask, z)
        self.umod = self.UpdateWeight(self.umod, upper, beta, delta, self.umask, z)





class HiddenLayer(CustomLayer):

    def __init__(self, feature_in, feature_out,feature_outout, has_Ymod=False, upper_ymod=True,change_scale_u=False,change_scale_v=False,higher_scale=False,Y1=False):
        super().__init__()
        K_size = 3
        self.LayerType = 'hidden'
        self.has_Ymod = has_Ymod # If the layer has a modulated group
        self.upper_ymod = upper_ymod # If the higher layer has a modulated group
        self.change_scale_u = change_scale_u
        self.change_scale_v = change_scale_v
        self.higher_scale = higher_scale
        self.Y1 = Y1
        if self.change_scale_v:
          self.v = nn.Conv2d(feature_in, feature_out, K_size, stride=3,bias=False)
          self.t = nn.Conv2d(feature_in, feature_out, K_size, stride=3, bias=False)
        else:
          self.v = nn.Conv2d(feature_in, feature_out, K_size, stride=1, padding='same',bias=False)
          if self.Y1:
             self.vmod = nn.Conv2d(feature_in, feature_out, K_size, stride=1, padding='same',bias=False)
          self.t = nn.Conv2d(feature_in, feature_out, K_size, stride=1, padding='same', bias=True)
        if has_Ymod or change_scale_u:
          if change_scale_u:
            self.u = nn.ConvTranspose2d(feature_outout, feature_out, 3, stride=3, padding=0,bias=False)
          else:
            self.u = nn.Conv2d(feature_outout, feature_out, K_size, stride=1, padding='same',bias=False)
          if upper_ymod:
              if change_scale_u:
                self.umod = nn.ConvTranspose2d(feature_outout, feature_out, 3, stride=3, padding=0,bias=False)
              else:
                self.umod = nn.Conv2d(feature_outout, feature_out, K_size, stride=1, padding='same', bias=False)

        # Initializing weights
        if self.change_scale_v:
          diag = False
        else:
          diag = True

        self.init_weights_FF(self.v,diag=True,change_scale = change_scale_v,higher_scale=higher_scale)
        if self.Y1:
          self.init_weights_FF(self.vmod,diag=True,change_scale = change_scale_v,higher_scale=higher_scale)
        self.init_weights_FF(self.t,diag=True,change_scale = change_scale_v,higher_scale=higher_scale)
        if has_Ymod or change_scale_u:
            self.init_weights_FB(self.u,change_scale = change_scale_u,higher_scale=higher_scale)
            if upper_ymod:
                self.init_weights_FB(self.umod,change_scale = change_scale_u,higher_scale=higher_scale)


        # Setting the mask to have connections only between neighboours
        self.vtmask = torch.tensor([[0, 0, 0], [0, 1, 0], [0, 0, 0]], requires_grad=False)
        if self.has_Ymod == False:
          self.vtmask = torch.tensor([[0, 0, 0], [0, 1, 0], [0, 0, 0]], requires_grad=False)
        self.vtmask = torch.tile(self.vtmask, (feature_out, feature_in, 1, 1))
        #if self.has_Ymod:
        #  for f in range(4):
        #    self.vtmask[f,0,:,:] = torch.tensor([[0, 1, 0], [1, 1, 1], [0, 1, 0]], requires_grad=False)

        self.tmask = torch.tensor([[0, 1, 0], [1, 1, 1], [0, 1, 0]], requires_grad=False)
        self.tmask = torch.tile(self.tmask, (feature_out, feature_in, 1, 1))

        self.vtmask = torch.clone(self.v.weight.detach())
        self.vtmask[self.vtmask != 0] = 1 

        self.tmask = torch.clone(self.t.weight.detach())
        self.tmask[self.tmask != 0] = 1         
 

        if has_Ymod or change_scale_u:
          self.umask = torch.clone(self.u.weight.detach())
          self.umask[self.umask != 0] = 1         

    def forward(self, lower_y=None, lower_ymod=None, upper_y=None, upper_ymod=None, det=False):
        if self.has_Ymod:
          current_y = self.ActivationFunction(self.v(lower_y))
        else:
          current_y = self.ActivationFunction(self.v(lower_y) + self.t(lower_ymod))
        return(current_y)

    def backward(self,same_y=None, lower_y=None, lower_ymod=None, upper_y=None, upper_ymod=None, det=False):
        if self.has_Ymod:
            if upper_ymod is not None:
                current_ymod = self.ActivationFunction((self.t(lower_ymod) + self.umod(upper_ymod)) *self.MultiFunction(same_y))
            else:
                current_ymod = self.ActivationFunction(self.MultiFunction(same_y) *(self.t(lower_ymod) + self.u(upper_y)))
        return(current_ymod)
        
    def UpdateLayer(self, upper, z, beta, delta):

        if self.change_scale_v:
          diag = False
        else:
          diag = True
        self.v = self.UpdateWeight(self.v, upper[0], beta, delta, self.vtmask, z[0],change_scale=self.change_scale_v)
        if not self.has_Ymod:
          self.t = self.UpdateWeight(self.t, upper[0], beta, delta, self.tmask, z[0],change_scale=self.change_scale_v)
        if self.has_Ymod:
            self.t = self.UpdateWeight(self.t, upper[1], beta, delta, self.tmask, z[1],change_scale=self.change_scale_v)
            if self.upper_ymod:
                self.umod = self.UpdateWeight(self.umod, upper[1], beta, delta, self.umask, z[1],change_scale = self.change_scale_u)
            else:
              self.u = self.UpdateWeight(self.u, upper[1], beta, delta, self.umask, z[1],change_scale = self.change_scale_u)

        #if self.change_scale_u:
            #self.u = self.UpdateWeight(self.u, upper[0], beta, delta, self.umask, z[0],change_scale = self.change_scale_u)

            #if self.upper_ymod:
            #    self.umod = self.UpdateWeight(self.umod, upper[0], beta, delta, self.umask, z[0],change_scale = self.change_scale_u)


class OutputLayer(CustomLayer):

    def __init__(self, hidden_features, input_features, feature_out,grid_size):
        super().__init__()
        self.grid_size = grid_size
        K_size = 2 * self.grid_size - 1
        self.LayerType = 'output'
        self.w = nn.ConvTranspose2d(6, 1,35, stride=3, padding=16,bias=True)#nn.ConvTranspose2d(6, feature_out, 3, stride=3, padding=0,bias=False)
        self.skipinterm = nn.Conv2d(4, feature_out, K_size, stride=1, padding='same',bias=True)
        self.skip = nn.Conv2d(input_features, feature_out, 1, stride=1, padding='same', bias=False)

        self.skipintermmask = (1/14) * torch.ones_like(self.skipinterm.weight)  # In numpy we do the average over 49*49-49 = 2352 whereas here over 13*13 -1 = 168 so we have to divide by 14 (number of strides)
        #self.wmask = torch.ones_like(self.w[0].weight,requires_grad = False)
        self.skipintermmask[:, :, self.grid_size - 1, self.grid_size - 1] = 1/50
        self.wmask = (1/14) * torch.ones_like(self.w.weight,requires_grad = False)
        self.wmask[:, :, 16:19,16:19] = 1/50



        # Initializing weights
        self.init_weights_FF(self.w)
        self.init_weights_FF(self.skip, disk=True) #not disk but same structure
        self.init_weights_FF(self.skipinterm,skip=True)


    def forward(self, lower_y,lower_lower_y, inputmod):
        Y = self.w(lower_y) + self.skip(inputmod)
        return(Y)

    def UpdateLayer(self, upper, beta, delta, winner_disk):
            self.w = self.UpdateWeight(self.w, upper, beta, delta, self.wmask)
            self.skip = self.UpdateWeight(self.skip, upper, beta, delta, average=False)
            #self.skipinterm = self.UpdateWeight(self.skipinterm, upper, beta, delta,self.skipintermmask, skip=True)




hide_toggle()
```

```python
# -*- coding: utf-8 -*-
"""
Created on Fri Oct 29 12:37:14 2021

@author: Sami
"""


class Network():

    def __init__(self, n_input_features,grid_size):
        self.beta = 0.02
        self.n_input_features = n_input_features
        self.n_hidden_features = 4

        self.exploitation_probability = 0.95

        self.grid_size = grid_size

        self.input_layer = InputLayer(self.n_input_features, 4)
        self.hidden_layer_1 = HiddenLayer(self.n_input_features, 4,4, has_Ymod=True,upper_ymod=True,Y1=False)
        self.hidden_layer_2 = HiddenLayer(self.n_hidden_features, self.n_hidden_features, 6,has_Ymod=True,change_scale_u=True)
        self.hidden_layer_3 = HiddenLayer(self.n_hidden_features,  6,6, has_Ymod=True,change_scale_v=True,higher_scale=True)
        self.hidden_layer_4 = HiddenLayer(6,6,6, has_Ymod=True,upper_ymod = False,higher_scale=True)
        self.hidden_layer_5 = HiddenLayer(6,6,6, has_Ymod=False,upper_ymod = False,higher_scale=True)
        self.output_layer = OutputLayer(self.n_hidden_features, self.n_input_features, 1,self.grid_size)

        self.save_activities = False
        
        self.saveX = []
        self.saveXmod = []
        self.saveY1 = []
        self.saveY1mod = []
        self.saveY2 = []
        self.saveY2mod = []
        self.saveY3 = []
        self.saveY3mod = []
        self.saveY4 = []
        self.saveY4mod = []
        self.saveY5 = []
        self.saveQ = []

        
    def doStep(self, input_env, reward, reset_traces, device):

        if self.save_activities:
            self.saveX.append([])
            self.saveXmod.append([])
            self.saveY1.append([])
            self.saveY1mod.append([])
            self.saveY2.append([])
            self.saveY2mod.append([])
            self.saveY3.append([])
            self.saveY3mod.append([])
            self.saveY4.append([])
            self.saveY4mod.append([])
            self.saveY5.append([])
            self.saveQ.append([])

        self.Xmod = torch.zeros(1, self.n_input_features, self.grid_size, self.grid_size, device=device)
        self.Y1 = torch.zeros(1, self.n_hidden_features, self.grid_size, self.grid_size, device=device)
        self.Y1mod = torch.zeros(1, self.n_hidden_features, self.grid_size, self.grid_size, device=device)
        self.Y2 = torch.zeros(1, self.n_hidden_features, self.grid_size, self.grid_size, device=device)
        self.Y2mod = torch.zeros(1, self.n_hidden_features, self.grid_size, self.grid_size, device=device)
        self.Y3 = torch.zeros(1, 6, self.grid_size//3, self.grid_size//3, device=device)
        self.Y3mod = torch.zeros(1, 6, self.grid_size//3, self.grid_size//3, device=device)
        self.Y4 = torch.zeros(1,6, self.grid_size//3, self.grid_size//3, device=device)
        self.Y4mod = torch.zeros(1, 6, self.grid_size//3, self.grid_size//3, device=device)
        self.Y5 = torch.zeros(1, 6, self.grid_size//3, self.grid_size//3, device=device)

        i = 0
        norm = 10
        flag = True

        while (i < 50 and norm > 0) or flag:

            prevY1mod = self.Y1mod.detach()
            (self.X) = self.input_layer.forward(self.Y1, self.Y1mod, input_env)
            (self.Y1) = self.hidden_layer_1.forward(lower_y=self.X, lower_ymod=self.Xmod, upper_y=self.Y2,upper_ymod=self.Y2mod)
            (self.Y2) = self.hidden_layer_2.forward(lower_y=self.Y1, lower_ymod=self.Y1mod, upper_y=self.Y3,upper_ymod=self.Y3mod)
            (self.Y3) = self.hidden_layer_3.forward(lower_y=self.Y2, lower_ymod=self.Y2mod, upper_y=self.Y4,upper_ymod=self.Y4mod)
            (self.Y4) = self.hidden_layer_4.forward(lower_y=self.Y3, lower_ymod=self.Y3mod, upper_y=self.Y5,upper_ymod=None)
            (self.Y5) = self.hidden_layer_5.forward(lower_y=self.Y4, lower_ymod=self.Y4mod, upper_y=None,upper_ymod=None)

            (self.Y4mod) = self.hidden_layer_4.backward(same_y=self.Y4,lower_y=self.Y3, lower_ymod=self.Y3mod, upper_y=self.Y5,upper_ymod=None)
            (self.Y3mod) = self.hidden_layer_3.backward(same_y=self.Y3,lower_y=self.Y2, lower_ymod=self.Y2mod, upper_y=self.Y4,upper_ymod=self.Y4mod)
            (self.Y2mod) = self.hidden_layer_2.backward(same_y=self.Y2,lower_y=self.Y1, lower_ymod=self.Y1mod, upper_y=self.Y3,upper_ymod=self.Y3mod)
            (self.Y1mod) = self.hidden_layer_1.backward(same_y=self.Y1,lower_y=self.X, lower_ymod=self.Xmod, upper_y=self.Y2,upper_ymod=self.Y2mod)
            (self.Xmod) = self.input_layer.backward(self.X,self.Y1, self.Y1mod, input_env)
            

            if self.save_activities:
                self.saveX[len(self.saveX) - 1].append(self.X.detach())
                self.saveXmod[len(self.saveXmod) - 1].append(self.Xmod.detach())
                self.saveY1[len(self.saveY1) - 1].append(self.Y1.detach())
                self.saveY1mod[len(self.saveY1mod) - 1].append(self.Y1mod.detach())
                self.saveY2[len(self.saveY2) - 1].append(self.Y2.detach())
                self.saveY2mod[len(self.saveY2mod) - 1].append(self.Y2mod.detach())
                self.saveY3[len(self.saveY3) - 1].append(self.Y3.detach())
                self.saveY3mod[len(self.saveY3mod) - 1].append(self.Y3mod.detach())
                self.saveY4[len(self.saveY4) - 1].append(self.Y4.detach())
                self.saveY4mod[len(self.saveY4mod) - 1].append(self.Y4mod.detach())
                self.saveY5[len(self.saveY5) - 1].append(self.Y5.detach())
                
                self.Z = self.calc_Output(device)
                self.saveQ[len(self.saveQ) - 1].append(self.Z.detach())

            with torch.no_grad():
                norm = torch.linalg.norm(self.Y1mod[0, :, :]-prevY1mod[0, :, :])

            i += 1
            if i == 6:
              flag = False

        ([self.Xmodtest, self.Y1test, self.Y1modtest, self.Y2test, self.Y2modtest,self.Y3test, self.Y3modtest,self.Y4test, self.Y4modtest,self.Y5test], [self.Xmod, self.Y1, self.Y1mod, self.Y2, self.Y2mod,self.Y3, self.Y3mod,self.Y4, self.Y4mod,self.Y5]) = self.detachAndreattach([self.Xmod, self.Y1, self.Y1mod, self.Y2, self.Y2mod,self.Y3, self.Y3mod,self.Y4, self.Y4mod,self.Y5])

        (self.X) = self.input_layer.forward(self.Y1, self.Y1mod, input_env)
        (self.Y1) = self.hidden_layer_1.forward(lower_y=self.X, lower_ymod=self.Xmod, upper_y=self.Y2,upper_ymod=self.Y2mod)
        (self.Y2) = self.hidden_layer_2.forward(lower_y=self.Y1, lower_ymod=self.Y1mod, upper_y=self.Y3,upper_ymod=self.Y3mod)
        (self.Y3) = self.hidden_layer_3.forward(lower_y=self.Y2, lower_ymod=self.Y2mod, upper_y=self.Y4,upper_ymod=self.Y4mod)
        (self.Y4) = self.hidden_layer_4.forward(lower_y=self.Y3, lower_ymod=self.Y3mod, upper_y=self.Y5,upper_ymod=None)
        (self.Y5) = self.hidden_layer_5.forward(lower_y=self.Y4, lower_ymod=self.Y4mod, upper_y=None,upper_ymod=None)

        (self.Y4mod) = self.hidden_layer_4.backward(same_y=self.Y4,lower_y=self.Y3, lower_ymod=self.Y3mod, upper_y=self.Y5,upper_ymod=None)
        (self.Y3mod) = self.hidden_layer_3.backward(same_y=self.Y3,lower_y=self.Y2, lower_ymod=self.Y2mod, upper_y=self.Y4,upper_ymod=self.Y4mod)
        (self.Y2mod) = self.hidden_layer_2.backward(same_y=self.Y2,lower_y=self.Y1, lower_ymod=self.Y1mod, upper_y=self.Y3,upper_ymod=self.Y3mod)
        (self.Y1mod) = self.hidden_layer_1.backward(same_y=self.Y1,lower_y=self.X, lower_ymod=self.Xmod, upper_y=self.Y2,upper_ymod=self.Y2mod)
        (self.Xmod) = self.input_layer.backward(self.X,self.Y1, self.Y1mod, input_env)

        self.Z = self.calc_Output(device)

        if self.save_activities:
            self.saveQ[len(self.saveQ) - 1].append(self.Z.detach())

        with torch.no_grad():
            action_chosen = torch.zeros((1, 1, 2+self.grid_size**2))
            action_chosen[0, 0, self.index_selected] = 1

        return (action_chosen)

    def calc_Output(self, device):
        Z = self.output_layer.forward(self.Y5,self.Y2, self.Xmod)
        with torch.no_grad():
            ZZ =torch.flatten(Z.permute(0,1,3,2), start_dim=2) #Flatten in F order beause everything is in F order
            if np.random.rand() < self.exploitation_probability:
                winner = self.calc_maxQ(ZZ)
            else:
                ZZint = ZZ.detach()
                ZZint -= torch.max(ZZint)
                ZZint = torch.exp(ZZint) / torch.sum(torch.exp(ZZint))
                winner = self.calc_softWTA(ZZint, device)
            self.index_selected = winner # index_selected represents the index of the action chosen on a flattened grid
            winner = [torch.tensor([0]), torch.tensor([0]), torch.tensor([(winner)%self.grid_size]), torch.tensor([torch.div(winner,self.grid_size,rounding_mode='floor')])]
            self.winner_disk = False
            self.action = winner # action represents the coordinates of the action chosen

        return Z

    def calc_maxQ(self, Z):
        winner = torch.where(Z == torch.max(Z))[-1]
        # Break ties randomly
        if len(winner) > 1:
            tiebreak = winner[torch.randint(0, len(winner), (1,))]
            winner = tiebreak
        return winner

    def calc_softWTA(self, probabilities, device):
        # Create wheel:
        probs = torch.cumsum(probabilities, 2)[0][0]

        # Select from wheel
        rnd = torch.rand((1,), device=device)
        for (i, prob) in enumerate(probs):
            if rnd <= prob:
                return i

    def Accessory(self):

        init = torch.autograd.grad(self.Z[self.action], [self.Y2, self.Y1mod, self.Y1, self.Xmod,self.Y2mod,self.Y3,self.Y3mod,self.Y4,self.Y4mod,self.Y5], retain_graph=True, allow_unused=True)

        Zy2 = init[0]
        Zy1mod = init[1]
        Zy1 = init[2]
        Zxmod = init[3]
        Zy2mod = init[4]
        Zy3 = init[5]
        Zy3mod = init[6]
        Zy4 = init[7]
        Zy4mod = init[8]
        Zy5 = init[9]

        for i in range(7):

            Zy5 = torch.autograd.grad(self.Y4mod, self.Y5, grad_outputs=Zy4mod, retain_graph=True, allow_unused=True)[0]
            Zy5 = Zy5 + init[9]

            Zy4mod = torch.autograd.grad(self.Y5, self.Y4modtest, grad_outputs=Zy5, retain_graph=True, allow_unused=True)[0]
            Zy4mod = Zy4mod + torch.autograd.grad(self.Y3mod, self.Y4mod, grad_outputs=Zy3mod, retain_graph=True, allow_unused=True)[0]
            Zy4mod = Zy4mod + init[8]

            Zy4 = torch.autograd.grad(self.Y5, self.Y4, grad_outputs=Zy5, retain_graph=True, allow_unused=True)[0]
            #Zy4 = Zy4 + torch.autograd.grad(self.Y3mod, self.Y4test, grad_outputs=Zy3mod, retain_graph=True, allow_unused=True)[0]
            Zy4 = Zy4 + init[7]

            Zy3mod = torch.autograd.grad(self.Y4mod, self.Y3modtest, grad_outputs=Zy4mod, retain_graph=True, allow_unused=True)[0]
            Zy3mod = Zy3mod + torch.autograd.grad(self.Y2mod, self.Y3mod, grad_outputs=Zy2mod, retain_graph=True, allow_unused=True)[0]
            Zy3mod = Zy3mod + init[6]

            Zy3 = torch.autograd.grad(self.Y4, self.Y3, grad_outputs=Zy4, retain_graph=True, allow_unused=True)[0]
            #Zy3 = Zy3 + torch.autograd.grad(self.Xmod, self.Y3test, grad_outputs=Zxmod, retain_graph=True, allow_unused=True)[0]
            Zy3 = Zy3 + init[5]

            Zy2mod = torch.autograd.grad(self.Y3mod, self.Y2modtest, grad_outputs=Zy3mod, retain_graph=True, allow_unused=True)[0]
            Zy2mod = Zy2mod + torch.autograd.grad(self.Y1mod, self.Y2mod, grad_outputs=Zy1mod, retain_graph=True, allow_unused=True)[0]
            Zy2mod = Zy2mod + init[4]

            Zy2 = torch.autograd.grad(self.Y3, self.Y2, grad_outputs=Zy3, retain_graph=True, allow_unused=True)[0]
            #Zy2 = Zy2 + torch.autograd.grad(self.Y1mod, self.Y2test, grad_outputs=Zy1mod, retain_graph=True, allow_unused=True)[0]
            Zy2 = Zy2 + init[0]

            Zy1mod = torch.autograd.grad(self.Y2mod, self.Y1modtest, grad_outputs=Zy2mod, retain_graph=True, allow_unused=True)[0]
            Zy1mod = Zy1mod + torch.autograd.grad(self.Xmod, self.Y1mod, grad_outputs=Zxmod, retain_graph=True, allow_unused=True)[0]
            Zy1mod = Zy1mod + init[1]

            Zy1 = torch.autograd.grad(self.Y2, self.Y1, grad_outputs=Zy2, retain_graph=True, allow_unused=True)[0]
            #Zy1 = Zy1 + torch.autograd.grad(self.Xmod, self.Y1test, grad_outputs=Zxmod, retain_graph=True, allow_unused=True)[0]
            Zy1 = Zy1 + init[2]

            Zxmod = torch.autograd.grad(self.Y1mod, self.Xmodtest, grad_outputs=Zy1mod, retain_graph=True, allow_unused=True)[0]
            Zxmod = Zxmod + init[3]

        return (Zxmod, Zy1, Zy1mod, Zy2,Zy2mod, Zy3,Zy3mod, Zy4,Zy4mod, Zy5)

    def doLearn(self, reward):
        with torch.no_grad():
            if self.winner_disk:
                exp_value = self.Z_disk[self.action]
            else:
                exp_value = self.Z[self.action]
            self.delta = reward - exp_value

        (Zxmod, Zy1, Zy1mod, Zy2,Zy2mod, Zy3,Zy3mod, Zy4,Zy4mod, Zy5) = self.Accessory()
        self.input_layer.UpdateLayer(self.Xmod, Zxmod, self.beta, self.delta)
        self.hidden_layer_1.UpdateLayer([self.Y1,self.Y1mod], [Zy1,Zy1mod], self.beta, self.delta)
        self.hidden_layer_2.UpdateLayer([self.Y2,self.Y2mod], [Zy2,Zy2mod], self.beta, self.delta)
        self.hidden_layer_3.UpdateLayer([self.Y3,self.Y3mod], [Zy3,Zy3mod], self.beta, self.delta)
        self.hidden_layer_4.UpdateLayer([self.Y4,self.Y4mod], [Zy4,Zy4mod], self.beta, self.delta)
        self.hidden_layer_5.UpdateLayer([self.Y5,self.Y5], [Zy5,Zy5], self.beta, self.delta)
        self.output_layer.UpdateLayer(self.Z[self.action], self.beta, self.delta, self.winner_disk)

    def detachAndreattach(self, x):
        detached = [xx.detach() for xx in x]
        for i, xx in enumerate(detached):
            xx.requires_grad = True
            x[i] = xx
        return(x, detached)

    def to(self, device):
        self.input_layer.to(device)
        self.hidden_layer_1.to(device)
        self.hidden_layer_2.toCusto(device)
        self.output_layer.to(device)
hide_toggle()
```

```python
def RunTrials(t,CurveLength,TrialNumber,n,device,dataset,index):
  n.exploit_prob = 1
  n.exploitation_probability = 1
  n.dotrain = False
  n.dosave = True
  n.save_activities = True
  corrects = []
  target_history = []
  distr_history = []
  feature_history = []
  display = []
  t.curve_length = CurveLength
  t.feature_target = 0
  action = np.zeros((1,49))
  action[0] = 1

  for p in range(TrialNumber):
    trial_running = True
    display.append([])
    feature_history.append(t.feature_target) 
    i=0
    new_input, reward, trialEnd= t.doStep(action)
    while trial_running:
      action = n.doStep(new_input,reward,trialEnd,device)
      new_input, reward, trialEnd = t.doStep(action)
      display[p].append(n.X.detach())
      if trialEnd:
        trial_running = False
        if reward == 0:
          corrects.append(0)
        else:
          corrects.append(1)
      i = i + 1
    target_history.append(t.target_curve)
    distr_history.append(t.distractor_curve) 
  #print(np.mean(corrects))
  return(n,corrects,feature_history,target_history,distr_history,display)  


hide_toggle()
```

```python
def model_latency(curve1,kernels,display,grid_size):
    
    stimulus = torch.sum(display,axis=(1)) 
    stimulus = stimulus[None,:]
    convolved = torch.zeros((1,6,6,6))
    convolved_reduced = torch.zeros((1,1,6,6))

    for i in range(len(kernels)):
        conv = torch.nn.functional.conv2d(stimulus, kernels[i], stride = 3)
        conv[conv < 0] = 0
        convolved[0,i,:,:] = conv
        convolved_reduced += conv
    
    
    L = []
    curve1Big = [((curve1[i] // grid_size) // 3)*6 + (curve1[i] % grid_size) // 3 for i in range(len(curve1))]

    current_count = 0
    last_kernel_saw = None
    i = 0
    while i < len(curve1Big):
        if np.where(convolved[:,:,curve1Big[i]%6,curve1Big[i]//6] != 0)[1].size == 0: #small scale
            current_count += 1
            L.append(current_count)
            i += 1
        else:
            if last_kernel_saw in np.where(convolved[:,:,curve1Big[i]%6,curve1Big[i]//6] != 0)[1]:
                current_count += 1
                L.append(current_count)
                while i < len(curve1Big)-1 and curve1Big[i] == curve1Big[i+1]:
                    L.append(current_count)
                    i += 1
                i += 1
            else:
                last_kernel_saw = np.where(convolved[:,:,curve1Big[i]%6,curve1Big[i]//6] != 0)[1]
                current_count += 1
                L.append(current_count)
                flag = False
                while i < len(curve1Big)-1 and curve1Big[i] == curve1Big[i+1]:
                    flag = True
                    L.append(current_count)# + 1)
                    i += 1
                if flag:
                    #current_count += 1
                    pass
                i += 1

    return(L)
```

```python
# -*- coding: utf-8 -*-
"""
Created on Fri Oct 29 12:29:28 2021

@author: Sami
"""
import numpy as np
import torch
import torch.nn as nn


class CustomLayer(nn.Module):

    def __init__(self):
        self.initialisation_range = 0.1
        super().__init__()

    def ActivationFunction(self, x):
        thresh = 6
        x[x == 0] = 0
        x = torch.clamp(x, 0)
        x[thresh <= x] = torch.log(1.5 * x[thresh <= x] + 1) + thresh - torch.log(torch.tensor(1.5 * thresh + 1))
        return x

    def MultiFunction(self,x):
      param = 100
      return param*x/(torch.sqrt(1+(param**2)*(x**2)))
      #return x

    def AverageTraces(self, traces, mask,skip=False,change_scale=False):
        traces *= mask
        if self.LayerType != "output":
            if change_scale:
                m = torch.mean(traces[:,:,:,:], axis=(2,3))
                m = m[:, :, None,None]
                traces[:,:] = m  
            else:
                m = torch.mean(traces[:, :, [0, 1, 1, 2], [1, 0, 2, 1]], axis=2)
                m = m[:, :, None]
                traces[:, :, [0, 1, 1, 2], [1, 0, 2, 1]] = m
        elif self.LayerType == "output" and not skip:
            traces *= mask
            intermmask = torch.ones((2 * self.grid_size - 1, 2 * self.grid_size - 1))
            intermmask[16:19,16:19] = 0
            m = torch.mean(traces[:, :, intermmask == 1], axis=2)
            traces[:, :, intermmask == 1] =  m[:,None]

            intermmask = torch.zeros((2 * self.grid_size - 1, 2 * self.grid_size - 1))
            intermmask[16:19,16:19] = 1
            m = torch.mean(traces[:, :, intermmask == 1], axis=2)
            traces[:, :, intermmask == 1] =  m[:,None]            
        elif self.LayerType == 'output' and skip:
            traces *= mask
            intermmask = torch.ones((2 * self.grid_size - 1, 2 * self.grid_size - 1))
            intermmask[self.grid_size - 1, self.grid_size - 1] = 0
            m = torch.mean(traces[:, :, intermmask == 1], axis=2)
            traces[:, :, intermmask == 1] = m.T          
        return(traces)

    def init_weights_FF(self, layer, disk=False,diag=False,change_scale=False,higher_scale = False,skip=False,rangee=1):
        K = torch.zeros(layer.weight.shape)
        if layer.bias is not None:
            bias = torch.zeros(layer.bias.shape)
        if change_scale:
          for f in range(K.shape[1]): 
            K[0,f,:,:] = torch.tensor([[[0.1*np.random.rand()+0.2,-1,-1],[0.1*np.random.rand()+0.2,-1,-1],[0.1*np.random.rand()+0.2,-1,-1]]])
            K[1,f,:,:] = torch.tensor([[[-1,0.1*np.random.rand()+0.2,-1],[-1,0.1*np.random.rand()+0.2,-1],[-1,0.1*np.random.rand()+0.2,-1]]])
            K[2,f,:,:] = torch.tensor([[[-1,-1,0.1*np.random.rand()+0.2],[-1,-1,0.1*np.random.rand()+0.2],[-1,-1,0.1*np.random.rand()+0.2]]])
            K[3,f,:,:] = torch.tensor([[[0.1*np.random.rand()+0.2,0.1*np.random.rand()+0.2,0.1*np.random.rand()+0.2],[-1,-1,-1],[-1,-1,-1]]])
            K[4,f,:,:] = torch.tensor([[[-1,-1,-1],[0.1*np.random.rand()+0.2,0.1*np.random.rand()+0.2,0.1*np.random.rand()+0.2],[-1,-1,-1]]])
            K[5,f,:,:] = torch.tensor([[[-1,-1,-1],[-1,-1,-1],[0.1*np.random.rand()+0.2,0.1*np.random.rand()+0.2,0.1*np.random.rand()+0.2]]])
          if layer.bias is not None:
              bias[f] = 0.1 * np.random.rand()
        if higher_scale and not change_scale:
          for f in range(K.shape[0]):
              for f2 in range(K.shape[1]):
                if f == f2 :
                      K[f, f2, 1, 1] = 0.1 * np.random.rand() + 0.2                                
        if not higher_scale:
          for f in range(K.shape[0]):
              for f2 in range(K.shape[1]):
                  if not disk:
                      if self.LayerType != "output":
                          K[f, f2, 1, 1] = self.initialisation_range * np.random.rand() * rangee
                      else:
                          if skip == False:
                              K[f, f2, :, :] = - 0.1 * np.random.rand() * torch.ones((K.shape[2], K.shape[3])) / 100
                              K[f, f2, 16:19,16:19] = 0.1 * np.random.rand()
                              #K[f, f2, :, :] = 0.1 * np.random.rand()
                          else:
                              K[f, f2, :, :] =  -0.1 * np.random.rand() * torch.ones((K.shape[2], K.shape[3])) / 100
                              K[f, f2, self.grid_size - 1, self.grid_size - 1] = 0.1 * np.random.rand()       
                  else:
                      K[f, f2, 0] = self.initialisation_range * np.random.rand()
              if layer.bias is not None:
                for f in range(bias.shape[0]):   
                  bias[f] = 0.1 * np.random.rand()
        layer.weight = torch.nn.Parameter(K)
        if layer.bias is not None:
            layer.bias = torch.nn.Parameter(bias)

    def init_weights_FB(self, layer, disk=False,diag=False,change_scale=False,higher_scale = False):
        K = torch.zeros(layer.weight.shape)
        if layer.bias is not None:
            bias = torch.zeros(layer.bias.shape)
        if higher_scale:
          for f in range(K.shape[0]):
              for f2 in range(K.shape[1]):
                if f == f2 :
                    if f2 in [0,1,2]:
                          K[f, f2, 1, 1] = 0.1 * np.random.rand()
                          #K[f, f2, [0, 1, 1, 2], [1, 0, 2, 1]] = 0.1 * np.random.rand()
                          K[f, f2, 0,1] = 0.1 * np.random.rand()+ 0.2
                          K[f, f2, 2,1] = 0.1 * np.random.rand()+ 0.2
                    elif f2 in [3,4,5]:
                          K[f, f2, 1, 1] = 0.1 * np.random.rand()
                          #K[f, f2, [0, 1, 1, 2], [1, 0, 2, 1]] = 0.1 * np.random.rand()
                          K[f, f2, 1,0] = 0.1 * np.random.rand()+ 0.2
                          K[f, f2, 1,2] = 0.1 * np.random.rand()+ 0.2                                                 
        else:                                            
          for f in range(K.shape[0]):
              for f2 in range(K.shape[1]):
                  if not disk:
                      if self.LayerType != "output":
                        if change_scale:
                          K[f, f2, :, :] = 0.1 * np.random.rand() + 0.2
                        else:
                            K[f, f2, 1, 1] = 0.1 * np.random.rand() 
                            #K[f, f2, [0, 1, 1, 2], [1, 0, 2, 1]] = 0.1 * np.random.rand()
                            K[f, f2, 0,1] = 0.1 * np.random.rand()+ 0.2
                            K[f, f2, 1,0] = 0.1 * np.random.rand()+ 0.2
                            K[f, f2, 1,2] = 0.1 * np.random.rand()+ 0.2
                            K[f, f2, 2,1] = 0.1 * np.random.rand()+ 0.2
                  else:
                      K[f, f2, 0] = self.initialisation_range * np.random.rand()
        if layer.bias is not None:
          for f in range(bias.shape[0]):   
            bias[f] = 0.1 * np.random.rand()
        layer.weight = torch.nn.Parameter(K)
        if layer.bias is not None:
            layer.bias = torch.nn.Parameter(bias)


    def UpdateWeight(self, layer, upper, beta, delta, mask=None, z=None, average=True,change_scale=False,skip=False):
        with torch.no_grad():
            
            # Getting the derivative of the output unit with respect to the 
            # weight and bias
            if layer.bias is not None:
                delta_weight = torch.autograd.grad(upper, [layer.weight, layer.bias], grad_outputs=z, retain_graph=True, allow_unused=True)
                delta_bias = delta_weight[1]
                delta_bias = torch.clamp(delta_bias, None, 1)
                #delta_bias[:] = torch.mean(delta_bias)
            else:
                delta_weight = torch.autograd.grad(upper, layer.weight, grad_outputs=z, retain_graph=True, allow_unused=True)
            delta_weight = delta_weight[0]
            delta_weight = torch.clamp(delta_weight, None, 1)
            
            # Averaging the weight and bias and updating the weights with RPE
            if average:
              if self.LayerType == 'output' or self.change_scale_u:
                delta_weight = self.AverageTraces(delta_weight, mask,change_scale=change_scale,skip=skip)
              else:
                delta_weight = delta_weight*mask
            weight_update = layer.weight + beta * delta * delta_weight
            layer.weight.copy_(weight_update)
            if layer.bias is not None:
                bias_update = layer.bias + beta * delta * delta_bias
                layer.bias.copy_(bias_update)
        return(layer)

    def to(self, device):
        for i in range(2):
            if self.LayerType == 'input':
                self.u[i].to(device)
                self.umod[i].to(device)
                self.umask = self.umask.to(device)
            elif self.LayerType == 'hidden':
                self.v[i].to(device)
                self.t[i].to(device)
                self.vtmask = self.vtmask.to(device)
                if self.has_Ymod:
                    self.u[i].to(device)
                    self.umask = self.umask.to(device)
                    if self.upper_ymod:
                        self.umod[i].to(device)
            elif self.LayerType == 'output':
                self.w[i].to(device)
                self.skip[i].to(device)
                self.wmask = self.wmask.to(device)
                self.skipmask = self.skipmask.to(device)
                
                
class InputLayer(CustomLayer):

    def __init__(self, feature_in, feature_out):
        super().__init__()
        self.LayerType = 'input'
        K_size = 3
        self.change_scale_u = False
        # First element is grid to grid, second element is disk to disk
        self.u = nn.Conv2d(feature_out, feature_in, K_size, stride=1, padding='same')
        self.umod = nn.Conv2d(feature_out, feature_in, K_size, stride=1, padding='same',bias=False)

        # Initializing weights
        self.init_weights_FB(self.u)
        self.init_weights_FB(self.umod)

        # Setting the mask to have connections only between neighboours
        self.umask = torch.tensor([[0, 1, 0], [1, 1, 1], [0, 1, 0]], requires_grad=False)
        self.umask = torch.tile(self.umask, (feature_in, feature_out, 1, 1))

    def forward(self, upper_y, upper_ymod, input):
        Y = input[0]
        return(Y)

    def backward(self,same_y, upper_y, upper_ymod, input):        
        Ymod = self.ActivationFunction(self.umod(upper_ymod) * self.MultiFunction(same_y))
        return(Ymod)

    def UpdateLayer(self, upper, z, beta, delta):
        #self.u= self.UpdateWeight(self.u, upper, beta, delta, self.umask, z)
        self.umod = self.UpdateWeight(self.umod, upper, beta, delta, self.umask, z)





class HiddenLayer(CustomLayer):

    def __init__(self, feature_in, feature_out,feature_outout, has_Ymod=False, upper_ymod=True,change_scale_u=False,change_scale_v=False,higher_scale=False,Y1=False):
        super().__init__()
        K_size = 3
        self.LayerType = 'hidden'
        self.has_Ymod = has_Ymod # If the layer has a modulated group
        self.upper_ymod = upper_ymod # If the higher layer has a modulated group
        self.change_scale_u = change_scale_u
        self.change_scale_v = change_scale_v
        self.higher_scale = higher_scale
        self.Y1 = Y1
        if self.change_scale_v:
          self.v = nn.Conv2d(feature_in, feature_out, K_size, stride=3,bias=False)
          self.t = nn.Conv2d(feature_in, feature_out, K_size, stride=3, bias=False)
        else:
          self.v = nn.Conv2d(feature_in, feature_out, K_size, stride=1, padding='same',bias=False)
          if self.Y1:
             self.vmod = nn.Conv2d(feature_in, feature_out, K_size, stride=1, padding='same',bias=False)
          self.t = nn.Conv2d(feature_in, feature_out, K_size, stride=1, padding='same', bias=True)
        if has_Ymod or change_scale_u:
          if change_scale_u:
            self.u = nn.ConvTranspose2d(feature_outout, feature_out, 3, stride=3, padding=0,bias=False)
          else:
            self.u = nn.Conv2d(feature_outout, feature_out, K_size, stride=1, padding='same',bias=False)
          if upper_ymod:
              if change_scale_u:
                self.umod = nn.ConvTranspose2d(feature_outout, feature_out, 3, stride=3, padding=0,bias=False)
              else:
                self.umod = nn.Conv2d(feature_outout, feature_out, K_size, stride=1, padding='same', bias=False)

        # Initializing weights
        if self.change_scale_v:
          diag = False
        else:
          diag = True

        self.init_weights_FF(self.v,diag=True,change_scale = change_scale_v,higher_scale=higher_scale)
        if self.Y1:
          self.init_weights_FF(self.vmod,diag=True,change_scale = change_scale_v,higher_scale=higher_scale)
        self.init_weights_FF(self.t,diag=True,change_scale = change_scale_v,higher_scale=higher_scale)
        if has_Ymod or change_scale_u:
            self.init_weights_FB(self.u,change_scale = change_scale_u,higher_scale=higher_scale)
            if upper_ymod:
                self.init_weights_FB(self.umod,change_scale = change_scale_u,higher_scale=higher_scale)


        # Setting the mask to have connections only between neighboours
        self.vtmask = torch.tensor([[0, 0, 0], [0, 1, 0], [0, 0, 0]], requires_grad=False)
        if self.has_Ymod == False:
          self.vtmask = torch.tensor([[0, 0, 0], [0, 1, 0], [0, 0, 0]], requires_grad=False)
        self.vtmask = torch.tile(self.vtmask, (feature_out, feature_in, 1, 1))
        #if self.has_Ymod:
        #  for f in range(4):
        #    self.vtmask[f,0,:,:] = torch.tensor([[0, 1, 0], [1, 1, 1], [0, 1, 0]], requires_grad=False)

        self.tmask = torch.tensor([[0, 1, 0], [1, 1, 1], [0, 1, 0]], requires_grad=False)
        self.tmask = torch.tile(self.tmask, (feature_out, feature_in, 1, 1))

        self.vtmask = torch.clone(self.v.weight.detach())
        self.vtmask[self.vtmask != 0] = 1 

        self.tmask = torch.clone(self.t.weight.detach())
        self.tmask[self.tmask != 0] = 1         
 

        if has_Ymod or change_scale_u:
          self.umask = torch.clone(self.u.weight.detach())
          self.umask[self.umask != 0] = 1         

    def forward(self, lower_y=None, lower_ymod=None, upper_y=None, upper_ymod=None, det=False):
        if self.has_Ymod:
          current_y = self.ActivationFunction(self.v(lower_y))
        else:
          current_y = self.ActivationFunction(self.v(lower_y) + self.t(lower_ymod))
        return(current_y)

    def backward(self,same_y=None, lower_y=None, lower_ymod=None, upper_y=None, upper_ymod=None, det=False):
        if self.has_Ymod:
            if upper_ymod is not None:
                current_ymod = self.ActivationFunction((self.t(lower_ymod) + self.umod(upper_ymod)) *self.MultiFunction(same_y))
            else:
                current_ymod = self.ActivationFunction(self.MultiFunction(same_y) *(self.t(lower_ymod) + self.u(upper_y)))
        return(current_ymod)
        
    def UpdateLayer(self, upper, z, beta, delta):

        if self.change_scale_v:
          diag = False
        else:
          diag = True
        if not self.change_scale_v:
          self.v = self.UpdateWeight(self.v, upper[0], beta, delta, self.vtmask, z[0],change_scale=self.change_scale_v)
          if not self.has_Ymod:
            self.t = self.UpdateWeight(self.t, upper[0], beta, delta, self.tmask, z[0],change_scale=self.change_scale_v)
        if self.has_Ymod:
            self.t = self.UpdateWeight(self.t, upper[1], beta, delta, self.tmask, z[1],change_scale=self.change_scale_v)
            if self.upper_ymod:
                self.umod = self.UpdateWeight(self.umod, upper[1], beta, delta, self.umask, z[1],change_scale = self.change_scale_u)
            else:
              self.u = self.UpdateWeight(self.u, upper[1], beta, delta, self.umask, z[1],change_scale = self.change_scale_u)

        #if self.change_scale_u:
            #self.u = self.UpdateWeight(self.u, upper[0], beta, delta, self.umask, z[0],change_scale = self.change_scale_u)

            #if self.upper_ymod:
            #    self.umod = self.UpdateWeight(self.umod, upper[0], beta, delta, self.umask, z[0],change_scale = self.change_scale_u)


class OutputLayer(CustomLayer):

    def __init__(self, hidden_features, input_features, feature_out,grid_size):
        super().__init__()
        self.grid_size = grid_size
        K_size = 2 * self.grid_size - 1
        self.LayerType = 'output'
        #self.w = nn.ConvTranspose2d(6, 1,35, stride=3, padding=16,bias=True)#nn.ConvTranspose2d(6, feature_out, 3, stride=3, padding=0,bias=False)
        self.w = nn.Conv2d(6, feature_out, K_size, stride=1, padding='same',bias=True)
        self.skipinterm = nn.Conv2d(4, feature_out, K_size, stride=1, padding='same',bias=True)
        self.skip = nn.Conv2d(input_features, feature_out, 1, stride=1, padding='same', bias=False)

        self.skipintermmask = (1/14) * torch.ones_like(self.skipinterm.weight)  # In numpy we do the average over 49*49-49 = 2352 whereas here over 13*13 -1 = 168 so we have to divide by 14 (number of strides)
        #self.wmask = torch.ones_like(self.w[0].weight,requires_grad = False)
        self.skipintermmask[:, :, self.grid_size - 1, self.grid_size - 1] = 1/50
        self.wmask = (1/14) * torch.ones_like(self.w.weight,requires_grad = False)
        self.wmask[:, :, self.grid_size - 1, self.grid_size - 1] = 1/50
        #self.wmask[:, :, 16:19,16:19] = 1/50



        # Initializing weights
        self.init_weights_FF(self.w,skip=True)
        self.init_weights_FF(self.skip, disk=True) #not disk but same structure
        self.init_weights_FF(self.skipinterm,skip=True)


    def forward(self, lower_y,lower_lower_y, inputmod):
        Y = self.w(lower_y) + self.skip(inputmod)
        return(Y)

    def UpdateLayer(self, upper, beta, delta, winner_disk):
            self.w = self.UpdateWeight(self.w, upper, beta, delta, self.wmask,skip=True)
            self.skip = self.UpdateWeight(self.skip, upper, beta, delta, average=False)
            #self.skipinterm = self.UpdateWeight(self.skipinterm, upper, beta, delta,self.skipintermmask, skip=True)


```

```python
# -*- coding: utf-8 -*-
"""
Created on Fri Oct 29 12:37:14 2021

@author: Sami
"""


class Network():

    def __init__(self, n_input_features,grid_size):
        self.beta = 0.02
        self.n_input_features = n_input_features
        self.n_hidden_features = 4

        self.exploitation_probability = 0.95

        self.grid_size = grid_size

        self.input_layer = InputLayer(self.n_input_features, 4)
        self.hidden_layer_1 = HiddenLayer(self.n_input_features, 4,4, has_Ymod=True,upper_ymod=True,Y1=False)
        self.hidden_layer_2 = HiddenLayer(self.n_hidden_features, self.n_hidden_features, 6,has_Ymod=True,change_scale_u=False)
        self.hidden_layer_3 = HiddenLayer(self.n_hidden_features,  6,6, has_Ymod=True,change_scale_v=False,higher_scale=False)
        self.hidden_layer_4 = HiddenLayer(6,6,6, has_Ymod=True,upper_ymod = False,higher_scale=False)
        self.hidden_layer_5 = HiddenLayer(6,6,6, has_Ymod=False,upper_ymod = False,higher_scale=False)
        self.output_layer = OutputLayer(self.n_hidden_features, self.n_input_features, 1,self.grid_size)

        self.save_activities = False
        
        self.saveX = []
        self.saveXmod = []
        self.saveY1 = []
        self.saveY1mod = []
        self.saveY2 = []
        self.saveY2mod = []
        self.saveY3 = []
        self.saveY3mod = []
        self.saveY4 = []
        self.saveY4mod = []
        self.saveY5 = []
        self.saveQ = []

        
    def doStep(self, input_env, reward, reset_traces, device):

        if self.save_activities:
            self.saveX.append([])
            self.saveXmod.append([])
            self.saveY1.append([])
            self.saveY1mod.append([])
            self.saveY2.append([])
            self.saveY2mod.append([])
            self.saveY3.append([])
            self.saveY3mod.append([])
            self.saveY4.append([])
            self.saveY4mod.append([])
            self.saveY5.append([])
            self.saveQ.append([])

        self.Xmod = torch.zeros(1, self.n_input_features, self.grid_size, self.grid_size, device=device)
        self.Y1 = torch.zeros(1, self.n_hidden_features, self.grid_size, self.grid_size, device=device)
        self.Y1mod = torch.zeros(1, self.n_hidden_features, self.grid_size, self.grid_size, device=device)
        self.Y2 = torch.zeros(1, self.n_hidden_features, self.grid_size, self.grid_size, device=device)
        self.Y2mod = torch.zeros(1, self.n_hidden_features, self.grid_size, self.grid_size, device=device)
        self.Y3 = torch.zeros(1, 6, self.grid_size, self.grid_size, device=device)
        self.Y3mod = torch.zeros(1, 6, self.grid_size, self.grid_size, device=device)
        self.Y4 = torch.zeros(1,6, self.grid_size, self.grid_size, device=device)
        self.Y4mod = torch.zeros(1, 6, self.grid_size, self.grid_size, device=device)
        self.Y5 = torch.zeros(1, 6, self.grid_size, self.grid_size, device=device)

        i = 0
        norm = 10
        flag = True

        while (i < 50 and norm > 0) or flag:

            prevY1mod = self.Y1mod.detach()
            (self.X) = self.input_layer.forward(self.Y1, self.Y1mod, input_env)
            (self.Y1) = self.hidden_layer_1.forward(lower_y=self.X, lower_ymod=self.Xmod, upper_y=self.Y2,upper_ymod=self.Y2mod)
            (self.Y2) = self.hidden_layer_2.forward(lower_y=self.Y1, lower_ymod=self.Y1mod, upper_y=self.Y3,upper_ymod=self.Y3mod)
            (self.Y3) = self.hidden_layer_3.forward(lower_y=self.Y2, lower_ymod=self.Y2mod, upper_y=self.Y4,upper_ymod=self.Y4mod)
            (self.Y4) = self.hidden_layer_4.forward(lower_y=self.Y3, lower_ymod=self.Y3mod, upper_y=self.Y5,upper_ymod=None)
            (self.Y5) = self.hidden_layer_5.forward(lower_y=self.Y4, lower_ymod=self.Y4mod, upper_y=None,upper_ymod=None)

            (self.Y4mod) = self.hidden_layer_4.backward(same_y=self.Y4,lower_y=self.Y3, lower_ymod=self.Y3mod, upper_y=self.Y5,upper_ymod=None)
            (self.Y3mod) = self.hidden_layer_3.backward(same_y=self.Y3,lower_y=self.Y2, lower_ymod=self.Y2mod, upper_y=self.Y4,upper_ymod=self.Y4mod)
            (self.Y2mod) = self.hidden_layer_2.backward(same_y=self.Y2,lower_y=self.Y1, lower_ymod=self.Y1mod, upper_y=self.Y3,upper_ymod=self.Y3mod)
            (self.Y1mod) = self.hidden_layer_1.backward(same_y=self.Y1,lower_y=self.X, lower_ymod=self.Xmod, upper_y=self.Y2,upper_ymod=self.Y2mod)
            (self.Xmod) = self.input_layer.backward(self.X,self.Y1, self.Y1mod, input_env)
            

            if self.save_activities:
                self.saveX[len(self.saveX) - 1].append(self.X.detach())
                self.saveXmod[len(self.saveXmod) - 1].append(self.Xmod.detach())
                self.saveY1[len(self.saveY1) - 1].append(self.Y1.detach())
                self.saveY1mod[len(self.saveY1mod) - 1].append(self.Y1mod.detach())
                self.saveY2[len(self.saveY2) - 1].append(self.Y2.detach())
                self.saveY2mod[len(self.saveY2mod) - 1].append(self.Y2mod.detach())
                self.saveY3[len(self.saveY3) - 1].append(self.Y3.detach())
                self.saveY3mod[len(self.saveY3mod) - 1].append(self.Y3mod.detach())
                self.saveY4[len(self.saveY4) - 1].append(self.Y4.detach())
                self.saveY4mod[len(self.saveY4mod) - 1].append(self.Y4mod.detach())
                self.saveY5[len(self.saveY5) - 1].append(self.Y5.detach())
                
                self.Z = self.calc_Output(device)
                self.saveQ[len(self.saveQ) - 1].append(self.Z.detach())

            with torch.no_grad():
                norm = torch.linalg.norm(self.Y1mod[0, :, :]-prevY1mod[0, :, :])

            i += 1
            if i == 6:
              flag = False

        ([self.Xmodtest, self.Y1test, self.Y1modtest, self.Y2test, self.Y2modtest,self.Y3test, self.Y3modtest,self.Y4test, self.Y4modtest,self.Y5test], [self.Xmod, self.Y1, self.Y1mod, self.Y2, self.Y2mod,self.Y3, self.Y3mod,self.Y4, self.Y4mod,self.Y5]) = self.detachAndreattach([self.Xmod, self.Y1, self.Y1mod, self.Y2, self.Y2mod,self.Y3, self.Y3mod,self.Y4, self.Y4mod,self.Y5])

        (self.X) = self.input_layer.forward(self.Y1, self.Y1mod, input_env)
        (self.Y1) = self.hidden_layer_1.forward(lower_y=self.X, lower_ymod=self.Xmod, upper_y=self.Y2,upper_ymod=self.Y2mod)
        (self.Y2) = self.hidden_layer_2.forward(lower_y=self.Y1, lower_ymod=self.Y1mod, upper_y=self.Y3,upper_ymod=self.Y3mod)
        (self.Y3) = self.hidden_layer_3.forward(lower_y=self.Y2, lower_ymod=self.Y2mod, upper_y=self.Y4,upper_ymod=self.Y4mod)
        (self.Y4) = self.hidden_layer_4.forward(lower_y=self.Y3, lower_ymod=self.Y3mod, upper_y=self.Y5,upper_ymod=None)
        (self.Y5) = self.hidden_layer_5.forward(lower_y=self.Y4, lower_ymod=self.Y4mod, upper_y=None,upper_ymod=None)

        (self.Y4mod) = self.hidden_layer_4.backward(same_y=self.Y4,lower_y=self.Y3, lower_ymod=self.Y3mod, upper_y=self.Y5,upper_ymod=None)
        (self.Y3mod) = self.hidden_layer_3.backward(same_y=self.Y3,lower_y=self.Y2, lower_ymod=self.Y2mod, upper_y=self.Y4,upper_ymod=self.Y4mod)
        (self.Y2mod) = self.hidden_layer_2.backward(same_y=self.Y2,lower_y=self.Y1, lower_ymod=self.Y1mod, upper_y=self.Y3,upper_ymod=self.Y3mod)
        (self.Y1mod) = self.hidden_layer_1.backward(same_y=self.Y1,lower_y=self.X, lower_ymod=self.Xmod, upper_y=self.Y2,upper_ymod=self.Y2mod)
        (self.Xmod) = self.input_layer.backward(self.X,self.Y1, self.Y1mod, input_env)

        self.Z = self.calc_Output(device)

        if self.save_activities:
            self.saveQ[len(self.saveQ) - 1].append(self.Z.detach())

        with torch.no_grad():
            action_chosen = torch.zeros((1, 1, 2+self.grid_size**2))
            action_chosen[0, 0, self.index_selected] = 1

        return (action_chosen)

    def calc_Output(self, device):
        Z = self.output_layer.forward(self.Y5,self.Y2, self.Xmod)
        with torch.no_grad():
            ZZ =torch.flatten(Z.permute(0,1,3,2), start_dim=2) #Flatten in F order beause everything is in F order
            if np.random.rand() < self.exploitation_probability:
                winner = self.calc_maxQ(ZZ)
            else:
                ZZint = ZZ.detach()
                ZZint -= torch.max(ZZint)
                ZZint = torch.exp(ZZint) / torch.sum(torch.exp(ZZint))
                winner = self.calc_softWTA(ZZint, device)
            self.index_selected = winner # index_selected represents the index of the action chosen on a flattened grid
            winner = [torch.tensor([0]), torch.tensor([0]), torch.tensor([(winner)%self.grid_size]), torch.tensor([torch.div(winner,self.grid_size,rounding_mode='floor')])]
            self.winner_disk = False
            self.action = winner # action represents the coordinates of the action chosen

        return Z

    def calc_maxQ(self, Z):
        winner = torch.where(Z == torch.max(Z))[-1]
        # Break ties randomly
        if len(winner) > 1:
            tiebreak = winner[torch.randint(0, len(winner), (1,))]
            winner = tiebreak
        return winner

    def calc_softWTA(self, probabilities, device):
        # Create wheel:
        probs = torch.cumsum(probabilities, 2)[0][0]

        # Select from wheel
        rnd = torch.rand((1,), device=device)
        for (i, prob) in enumerate(probs):
            if rnd <= prob:
                return i

    def Accessory(self):

        init = torch.autograd.grad(self.Z[self.action], [self.Y2, self.Y1mod, self.Y1, self.Xmod,self.Y2mod,self.Y3,self.Y3mod,self.Y4,self.Y4mod,self.Y5], retain_graph=True, allow_unused=True)

        Zy2 = init[0]
        Zy1mod = init[1]
        Zy1 = init[2]
        Zxmod = init[3]
        Zy2mod = init[4]
        Zy3 = init[5]
        Zy3mod = init[6]
        Zy4 = init[7]
        Zy4mod = init[8]
        Zy5 = init[9]

        for i in range(7):

            Zy5 = torch.autograd.grad(self.Y4mod, self.Y5, grad_outputs=Zy4mod, retain_graph=True, allow_unused=True)[0]
            Zy5 = Zy5 + init[9]

            Zy4mod = torch.autograd.grad(self.Y5, self.Y4modtest, grad_outputs=Zy5, retain_graph=True, allow_unused=True)[0]
            Zy4mod = Zy4mod + torch.autograd.grad(self.Y3mod, self.Y4mod, grad_outputs=Zy3mod, retain_graph=True, allow_unused=True)[0]
            Zy4mod = Zy4mod + init[8]

            Zy4 = torch.autograd.grad(self.Y5, self.Y4, grad_outputs=Zy5, retain_graph=True, allow_unused=True)[0]
            #Zy4 = Zy4 + torch.autograd.grad(self.Y3mod, self.Y4test, grad_outputs=Zy3mod, retain_graph=True, allow_unused=True)[0]
            Zy4 = Zy4 + init[7]

            Zy3mod = torch.autograd.grad(self.Y4mod, self.Y3modtest, grad_outputs=Zy4mod, retain_graph=True, allow_unused=True)[0]
            Zy3mod = Zy3mod + torch.autograd.grad(self.Y2mod, self.Y3mod, grad_outputs=Zy2mod, retain_graph=True, allow_unused=True)[0]
            Zy3mod = Zy3mod + init[6]

            Zy3 = torch.autograd.grad(self.Y4, self.Y3, grad_outputs=Zy4, retain_graph=True, allow_unused=True)[0]
            #Zy3 = Zy3 + torch.autograd.grad(self.Xmod, self.Y3test, grad_outputs=Zxmod, retain_graph=True, allow_unused=True)[0]
            Zy3 = Zy3 + init[5]

            Zy2mod = torch.autograd.grad(self.Y3mod, self.Y2modtest, grad_outputs=Zy3mod, retain_graph=True, allow_unused=True)[0]
            Zy2mod = Zy2mod + torch.autograd.grad(self.Y1mod, self.Y2mod, grad_outputs=Zy1mod, retain_graph=True, allow_unused=True)[0]
            Zy2mod = Zy2mod + init[4]

            Zy2 = torch.autograd.grad(self.Y3, self.Y2, grad_outputs=Zy3, retain_graph=True, allow_unused=True)[0]
            #Zy2 = Zy2 + torch.autograd.grad(self.Y1mod, self.Y2test, grad_outputs=Zy1mod, retain_graph=True, allow_unused=True)[0]
            Zy2 = Zy2 + init[0]

            Zy1mod = torch.autograd.grad(self.Y2mod, self.Y1modtest, grad_outputs=Zy2mod, retain_graph=True, allow_unused=True)[0]
            Zy1mod = Zy1mod + torch.autograd.grad(self.Xmod, self.Y1mod, grad_outputs=Zxmod, retain_graph=True, allow_unused=True)[0]
            Zy1mod = Zy1mod + init[1]

            Zy1 = torch.autograd.grad(self.Y2, self.Y1, grad_outputs=Zy2, retain_graph=True, allow_unused=True)[0]
            #Zy1 = Zy1 + torch.autograd.grad(self.Xmod, self.Y1test, grad_outputs=Zxmod, retain_graph=True, allow_unused=True)[0]
            Zy1 = Zy1 + init[2]

            Zxmod = torch.autograd.grad(self.Y1mod, self.Xmodtest, grad_outputs=Zy1mod, retain_graph=True, allow_unused=True)[0]
            Zxmod = Zxmod + init[3]

        return (Zxmod, Zy1, Zy1mod, Zy2,Zy2mod, Zy3,Zy3mod, Zy4,Zy4mod, Zy5)

    def doLearn(self, reward):
        with torch.no_grad():
            if self.winner_disk:
                exp_value = self.Z_disk[self.action]
            else:
                exp_value = self.Z[self.action]
            self.delta = reward - exp_value

        (Zxmod, Zy1, Zy1mod, Zy2,Zy2mod, Zy3,Zy3mod, Zy4,Zy4mod, Zy5) = self.Accessory()
        self.input_layer.UpdateLayer(self.Xmod, Zxmod, self.beta, self.delta)
        self.hidden_layer_1.UpdateLayer([self.Y1,self.Y1mod], [Zy1,Zy1mod], self.beta, self.delta)
        self.hidden_layer_2.UpdateLayer([self.Y2,self.Y2mod], [Zy2,Zy2mod], self.beta, self.delta)
        self.hidden_layer_3.UpdateLayer([self.Y3,self.Y3mod], [Zy3,Zy3mod], self.beta, self.delta)
        self.hidden_layer_4.UpdateLayer([self.Y4,self.Y4mod], [Zy4,Zy4mod], self.beta, self.delta)
        self.hidden_layer_5.UpdateLayer([self.Y5,self.Y5], [Zy5,Zy5], self.beta, self.delta)
        self.output_layer.UpdateLayer(self.Z[self.action], self.beta, self.delta, self.winner_disk)

    def detachAndreattach(self, x):
        detached = [xx.detach() for xx in x]
        for i, xx in enumerate(detached):
            xx.requires_grad = True
            x[i] = xx
        return(x, detached)

    def to(self, device):
        self.input_layer.to(device)
        self.hidden_layer_1.to(device)
        self.hidden_layer_2.toCusto(device)
        self.output_layer.to(device)
```
